```python
import json
```

```python
f = open("our_weather.json", "rb")
js = json.load(f)
```

```python
js[0]["m_weather_forecast_samples"][0].keys()
```

```python
relational_data = []
no_forecast = []
for i in range(len(js)):
    print(i)
    if "m_weather_forecast_samples" in js[i]:
        for j in range(len(js[i]["m_weather_forecast_samples"])):
            if js[i]["m_weather_forecast_samples"][j]["m_session_type"] == js[i]["m_weather_forecast_sample_m_session_type"]:
                doc = js[i].copy()
                doc.pop("m_weather_forecast_samples")
                doc['m_weather_forecast_samples_m_session_type'] = js[i]['m_weather_forecast_samples'][j]["m_weather_forecast_sample_m_session_type"]
                doc["m_time_offset"] = js[i]['m_weather_forecast_samples'][j]["m_time_offset"]
                doc["m_weather_forecast_sample_m_weather"] = js[i]['m_weather_forecast_samples'][j]["m_weather_forecast_sample_m_weather"]
                doc["m_track_temperature"] = js[i]['m_weather_forecast_samples'][j]["m_track_temperature"]
                doc["m_air_temperature"] = js[i]['m_weather_forecast_samples'][j]["m_air_temperature"]
                relational_data.append(doc)
    else:
        no_forecast.append(i)
```

```python
len(no_forecast)
```

```python
import pandas as pd
df = pd.DataFrame(relational_data)
df.head(50)
```

```python
df.columns = [df.columns[i].upper() for i in range(len(df.columns))]
```

```python
df.to_csv("weather_json.csv")
```
