```python
import json
```

```python
f = open("weather.json", "rb")
js = json.load(f)
```

```python
for i in range(len(js)):
    js[i].pop("m_marshal_zones")
```

```python
forecast= []
for i in range(len(js)):
    for j in range(len(js[i]["m_weather_forecast_samples"])):
        forecast.append(js[i]["m_weather"])
```

```python
import numpy as np
arr = np.array(forecast)
np.unique(arr, return_counts = True)
```

```python
for i in range(len(js)):
    for j in js[i]["m_header"]:
        js[i][j] = js[i]["m_header"][j]
        
    js[i].pop("m_header")
```

```python
relational_data = []
no_forecast = []
for i in range(len(js)):
    if i% 10000 == 0:
        print(i)
    if "m_weather_forecast_samples" in js[i]:
        for j in range(len(js[i]["m_weather_forecast_samples"])):
            if js[i]["m_weather_forecast_samples"][j]["m_session_type"] == js[i]["m_session_type"]:
                doc = js[i].copy()
                doc.pop("m_weather_forecast_samples")
                doc['m_weather_forecast_samples_m_session_type'] = js[i]['m_weather_forecast_samples'][j]["m_session_type"]
                doc["m_time_offset"] = js[i]['m_weather_forecast_samples'][j]["m_time_offset"]
                doc["m_weather_forecast_samples_m_weather"] = js[i]['m_weather_forecast_samples'][j]["m_weather"]
                doc["m_track_temperature"] = js[i]['m_weather_forecast_samples'][j]["m_track_temperature"]
                doc["m_track_temperature_change"] = js[i]['m_weather_forecast_samples'][j]["m_track_temperature_change"]
                doc["m_air_temperature"] = js[i]['m_weather_forecast_samples'][j]["m_air_temperature"]
                doc["m_air_temperature_change"] = js[i]['m_weather_forecast_samples'][j]["m_air_temperature_change"]
                doc["m_rain_percentage"] = js[i]['m_weather_forecast_samples'][j]["m_rain_percentage"]
                relational_data.append(doc)
    else:
        no_forecast.append(i)
```

```python
import pandas as pd
df = pd.DataFrame(relational_data)
df.head(50)
```

```python
df.columns = [df.columns[i].upper() for i in range(len(df.columns))]
```

```python
df.to_csv("weather_json.csv")
```
